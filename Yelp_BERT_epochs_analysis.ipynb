## Conclusions

After observing the results by averages we can conclude that the best parameters are **maximum length of 128 and batch size of 32**. 

With these parameters for the BertForSequentialClassification model we obtain the smallest Validation Loss and the best Validation Accuracy.

As for the number of epochs, we can also observe that by running the model for 3 epochs we obtain very similar results in Validation Accuracy and we can have more similar loss between the Tranining and Validation sets than with 10 epochs.
